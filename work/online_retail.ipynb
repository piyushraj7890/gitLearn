## Task 1: Load the Data

```python
# your code here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
online_retail_data = pd.read_excel('Online Retail.xlsx')
online_retail_data.head(10)
```

```python
online_retail_data.describe(include='all')
```

```python
online_retail_data.isna().sum()
```

```python
online_retail_data.dropna(inplace=True)
online_retail_data.isna().sum()
```

```python
online_retail_data.head(10)
```

```python
online_retail_data.count()
```

```python
duplicated_df = online_retail_data[online_retail_data.duplicated()]
duplicated_df.count()
```

```python
df = duplicated_df.drop_duplicates()
df.count()
print(df.isna().sum())
```

```python
df.head()
```

```python
from scipy.stats import zscore
df['Z-score UnitPrice'] = zscore(df['UnitPrice'])
z_score_threshold = 3
outliers = df[abs(df['Z-score UnitPrice']) > z_score_threshold]


# # Display the outliers
print("Potential Outliers:")
print(outliers)
# Remove the outliers from the dataset
df_no_outliers = df[abs(df['Z-score UnitPrice']) <= z_score_threshold]

# Display the dataset without outliers
print("Dataset without Outliers:")
print(df_no_outliers)
```

```python
df['Description'].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Retail")
plt.ylabel('Count')
plt.xlabel('Product');
```

```python
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Quantity'], bins=10, color='skyblue')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.title('Histogram of Quantity')
plt.grid(True)
plt.show()

# Scatter plot of Z-score UnitPrice against Quantity
plt.figure(figsize=(8, 6))
plt.scatter(df['Quantity'], df['Z-score UnitPrice'], color='orange', alpha=0.8)
plt.xlabel('Quantity')
plt.ylabel('Z-score UnitPrice')
plt.title('Scatter Plot: Quantity vs. Z-score UnitPrice')
plt.grid(True)
plt.show()
```

```python

# Convert 'InvoiceDate' column to datetime type
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract month and day of the week from 'InvoiceDate'
df.loc[:, 'Month'] = df['InvoiceDate'].dt.month
df.loc[:, 'DayOfWeek'] = df['InvoiceDate'].dt.dayofweek

# Group by month and day of the week and calculate total sales for each group
monthly_sales = df.groupby('Month')['Quantity'].sum()

# Create a new DataFrame to store the sales data for each day of the week
daily_sales_data = df.groupby('DayOfWeek')['Quantity'].sum()
daily_sales = pd.DataFrame({'DayOfWeek': daily_sales_data.index, 'TotalSales': daily_sales_data.values})

# Plotting the sales trends over time
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trends')
plt.grid(True)
plt.xticks(range(1, 13))
plt.show()

# Plotting the busiest days of the week in terms of sales
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(8, 6))
plt.bar(daily_sales['DayOfWeek'], daily_sales['TotalSales'])
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.title('Sales by Day of the Week')
plt.grid(True)
plt.show()
```

```python
top_selling_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

# Plotting the top-selling products
plt.figure(figsize=(12, 6))
plt.bar(top_selling_products.index, top_selling_products.values)
plt.xlabel('Product Description')
plt.ylabel('Total Quantity Sold')
plt.title('Top-Selling Products')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()

# Group by 'Country' and calculate total quantity sold for each country
top_selling_countries = df.groupby('Country')['Quantity'].sum().sort_values(ascending=False).head(10)

# Plotting the top-selling countries
plt.figure(figsize=(12, 6))
plt.bar(top_selling_countries.index, top_selling_countries.values)
plt.xlabel('Country')
plt.ylabel('Total Quantity Sold')
plt.title('Top-Selling Countries')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```
